```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = "model\dqn\statistics\dqn_cumulative_reward.txt"  # Replace with your file path
data = []

with open(file_path, "r") as file:
    for line in file:
        data.append(float(line.strip()))  # Convert each line to a float and add to the list

print(max(data))
print(np.mean(data))
plt.figure(figsize=(20, 6))
plt.plot(data)

plt.title("DQN - Reward per episode")
plt.xlabel("Episode")
plt.ylabel("Reward")


plt.show()
```

```python
file_path = "model\ppo\statistics\ppo_cumulative_reward.txt"  # Replace with your file path
data = []

with open(file_path, "r") as file:
    for line in file:
        data.append(float(line.strip()))  # Convert each line to a float and add to the list

print(max(data))
print(np.argmax(data))
print(np.mean(data))
plt.figure(figsize=(20, 6))
plt.plot(data)

plt.title("DQN - Reward per episode")
plt.xlabel("Episode")
plt.ylabel("Reward")


plt.show()
```
